```python
import yfinance as yf
import matplotlib.pyplot as plt

# Simple graph function
def make_graph(stock_data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(stock_data['Date'], stock_data['Close'], color='purple')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.grid(True)
    plt.savefig('gme_stock.png')
    plt.show()

# Get GameStop data
try:
    gme = yf.Ticker("GME")
    gme_data = gme.history(period="1y").reset_index()
    make_graph(gme_data, "GameStop Stock Price (Last Year)")
    
except Exception as e:
    print("Error occurred:", e)
```
